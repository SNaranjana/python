```python
import pandas as pd

# Create sales data
data = {
    "Order ID": [1001, 1002, 1003, 1004, 1005],
    "Customer Name": ["John", "Mary", "David", "Priya", "Arun"],
    "Retail Price": [45.99, 25.99, 99.99, 79.99, 149.99],
    "Order Quantity": [2, 1, 3, 2, 1]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Calculate Tax (10%)
df["Tax (10%)"] = df["Retail Price"] * df["Order Quantity"] * 0.10

# Calculate Total
df["Total Amount"] = (df["Retail Price"] * df["Order Quantity"]) + df["Tax (10%)"]

# Display Data
print("Supermarket Sales Data\n")
print(df)
```
